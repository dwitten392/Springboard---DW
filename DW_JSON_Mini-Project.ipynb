```python
#read into DF and preview the DF
path = 'data_wrangling_json\data_wrangling_json\data\world_bank_projects.json'
world_bank_df = pd.read_json(path)
world_bank_df.head(3)
```

```python
#separate out the mjtheme_name column into multiple columns of a new dataframe (for efficiency)
new_df = pd.DataFrame(world_bank_df['mjtheme_namecode'].values.tolist(), columns=['code1','code2', 'code3', 'code4', 'code5'])
new_df.head()
```

Count for each code (Top 10 only)

```python
# appends all codes in all columns to an empty list
m = []

for column in new_df.columns:
    for d in new_df[column]:
        try:
            m.append(d['code'])
        except:
            pass
        
#converts the list to a dictionary where the code is the key and the value is the count
d = {x:m.count(x) for x in m}

#creating a new dataframe with codes and counts
new_df1 = pd.DataFrame()
new_df1['code'] = d.keys()
new_df1['count'] = d.values()
new_df1 = new_df1.set_index('code')
print(new_df1.sort_values(by=['count'], ascending = False).head(10))
    
    
```

Dataframe with the code_name and the count

```python
# create a new dictionary with code as key and name as value
# very ugly and manual
code_name_dict = {'1': 'Economic management', '2' : 'Public sector governance', '3': 'Rule of law', '4' : 'Financial and private sector development', '5' : 'Trade and integration', '6' : 'Social protection and risk management', '7': 'Social dev/gender/inclusion', '8': 'Human development', '9': 'Urban development', '10': 'Rural development' , '11' : 'Environment and natural resources management'}
new_df2 = pd.DataFrame.from_dict(code_name_dict, orient = 'index', columns = ['code_name'])
new_df3 = new_df2.merge(new_df1, left_index = True, right_index = True)
new_df3.index.name = 'code'
print(new_df3.sort_values(by=['count'], ascending = False))
```
